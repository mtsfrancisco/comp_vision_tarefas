```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import label
```

```python
image = cv2.imread("../images/coins1.jpg")
image = cv2.medianBlur(image, 5)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(image, 100, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

plt.imshow(binary_image, cmap='gray')
```

```python
kernel = np.ones((5,5),np.uint8)
binary_image = cv2.dilate(binary_image, kernel, iterations=2)


plt.imshow(binary_image, cmap="gray")
```

```python
structure = np.ones((3, 3), dtype=int)

rotulos, num_objects = label(binary_image, structure=structure)
""""
O objetivo é identificar e agrupar regiões conectadas de pixels que possuem o mesmo valor (tipicamente 1).
Structure -> define a conectividade dos pixels

Flood Fill
"""

plt.imshow(rotulos, cmap="jet")
plt.show()

print(f"Número de objetos encontrados: {num_objects}")

```
